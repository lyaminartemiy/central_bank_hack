# CatBoost Baseline

```python
data_train.shape
```

```python
data_train.columns
```

```python
date_columns = ['Крайний срок', 'Дата обращения', 'Дата закрытия обращения', 'Дата восстановления']
data_train[['Крайний срок', 'Дата обращения', 'Дата закрытия обращения']]
```

```python
data_train = data_train.drop(columns=date_columns).dropna()
```

```python
data_train.columns
```

```python
target_columns = ['Тип переклассификации', 'Тип обращения итоговый']
X_train, X_valid, y_train, y_valid = train_test_split(data_train.drop(columns=target_columns), data_train['Тип обращения итоговый'], test_size=0.2, random_state=17)
X_train.shape, X_valid.shape
```

```python
X_train.dtypes
```

```python
X_train.isna().sum()
```

```python
cat_cols = list(X_train.columns)
```

```python
from catboost import Pool, cv, CatBoostClassifier

model = CatBoostClassifier(
    random_seed=63,
    loss_function='Logloss',
    iterations=1000,
    learning_rate=0.05,
    task_type='CPU',
    eval_metric='F1',
)

model.fit(
    Pool(data=X_train, label=y_train, cat_features=cat_cols),
    verbose=50,
    eval_set=Pool(data=X_valid, label=y_valid, cat_features=cat_cols),
    plot=True
)
```
